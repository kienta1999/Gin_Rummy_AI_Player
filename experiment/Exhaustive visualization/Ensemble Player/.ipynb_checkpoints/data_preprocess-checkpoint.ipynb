```python
import pandas as pd
```

```python
def mergeDF(df1, df2):
    winRate1 = df1['winRate']
    winRate2 = df2['winRate']
    winRate = (winRate1 + winRate2) / 2
    df = df1.copy()
    df['winRate'] = winRate
    return df
```

```python
pd.read_csv('raw data/Iteration 3/it3-6E-run2.csv')
```

```python
iteration = ['2', '3']
experiment = ['1', '2', '3', '4', '6']
player = ['E']
run = ['1', '2']
```

```python
for i in iteration:
    for e in experiment:
        for p in player:
            path1 = 'raw data/Iteration ' + i + '/it' + i + '-' + e + p + '-run' + '1' + '.csv'
            path2 = 'raw data/Iteration ' + i + '/it' + i + '-' + e + p + '-run' + '2' + '.csv'
            data1 = pd.read_csv(path1)
            data2 = pd.read_csv(path2)
            data = mergeDF(data1, data2)
            path = 'it' + i + '-' + e + p + '.csv'
            print(path1, path2, path)
            data.to_csv(path, index = False)
```
