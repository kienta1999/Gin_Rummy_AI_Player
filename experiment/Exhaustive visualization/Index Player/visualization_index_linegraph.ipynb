```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

winRate plot

```python
iteration = ['1', '2', '3']
experiment = ['1', '2', '3', '4', '6']
```

```python
data = {}
worst_win_rate = {}
mean_win_rate = {}
best_win_rate = {}
z_score = {}

id_to_experiment = {}
count = 0
for i in iteration:
    for e in experiment:
        path = 'it' + i + '-' + e + 'I.csv'
        key = i + e
        id_to_experiment[count] = key
        count += 1
        df = pd.read_csv(path)
        data[key] = df
        worst_win_rate[key] = df['winRate'].min()
        mean_win_rate[key] = df['winRate'].mean()
        best_win_rate[key] = df['winRate'].max()
        z_score[key] = (best_win_rate[key] - mean_win_rate[key]) / df['winRate'].std()
```

```python
x_axis = [i for i in range(len(data))] #id of experiment
x_axis
```

```python
id_to_experiment
```

```python
plt.plot(x_axis, [best_win_rate[i] for i in best_win_rate], label = "Best")
plt.plot(x_axis, [mean_win_rate[i] for i in mean_win_rate], label = "Mean")
plt.plot(x_axis, [worst_win_rate[i] for i in worst_win_rate], label = "Worst")

plt.xlabel('experiment')
plt.ylabel('winRate') 

plt.legend()

plt.title('Graph of experiment vs winRate') 
plt.show()
```

z-score plot

```python
z_score
```

```python
plt.plot(x_axis, [z_score[i] for i in z_score], label = "Z-score")

plt.xlabel('experiment')
plt.ylabel('z-score') 

plt.legend()

plt.title('Graph of experiment vs highest z-scorebest_win_rate') 
plt.show()
```
