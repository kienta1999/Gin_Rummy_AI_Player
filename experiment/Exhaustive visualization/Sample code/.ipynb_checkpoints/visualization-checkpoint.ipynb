```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
sns.set()
```

```python
def nD_to_2D(df, chosenColumns):
    c0 = chosenColumns[0]
    c1 = chosenColumns[1]
    param0 = df[c0].unique()
    param1 = df[c1].unique()
    df_2D = pd.DataFrame(columns = [c0, c1, 'winRate'])
    for p0 in param0:
        for p1 in param1:
            data = df[(df[c0] == p0) & (df[c1] == p1)]
            if data.shape[0] != 0:
                winRate = sum(data['winRate']) / data.shape[0]
            else:
                winRate = 0
            df_2D.loc[len(df_2D)] = [p0, p1, winRate]
    return df_2D
```

```python
def df2D_to_heatmap(df_2D, index_2d, columns_2d, values_2d):
    df_2D = df_2D.pivot(index=index_2d, columns=columns_2d, values=values_2d)
    ax = sns.heatmap(df_2D, cmap='gray')
    return ax
```

# MeldabilityHE

```python
meldability_params_data = pd.read_csv("ex1-1.csv")
meldability_params_data.head()
```

```python
meldability_params_data['p3'] = round(meldability_params_data['p3'], 2)
meldability_params_data['p4'] = round(meldability_params_data['p4'], 2)
meldability_params_data['p5'] = round(meldability_params_data['p5'], 2)
meldability_params_data['p6'] = round(meldability_params_data['p6'], 2)
```

### Heatmap

0 <= p3 <=  0.5; step = 0.05
0 <= p4 <=  1; step = 0.1
0 <= p5 <= 1; step = 0.1
0 <= p6 <= 1; step = 0.1

```python
data_p3_p4 = nD_to_2D(meldability_params_data, ['p3', 'p4'])
df2D_to_heatmap(data_p3_p4, 'p3', 'p4', 'winRate')
```

### 3D surface (color map)

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(meldability_params_data['p4'], meldability_params_data['p5'], meldability_params_data['p6'], marker=m, c = meldability_params_data['winRate'])

ax.set_xlabel('p4')
ax.set_ylabel('p5')
ax.set_zlabel('p6')

legend1 = ax.legend(*scatter.legend_elements(num=5),
                    loc="upper left", title="winrate")
ax.add_artist(legend1)

plt.show()
```
